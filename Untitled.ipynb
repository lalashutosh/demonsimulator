```python
import numpy as np
import matplotlib.pyplot as plt
from collections import namedtuple
```

```python
# Define particle properties
Particle = namedtuple('Particle', ['id', 'mass', 'pos', 'vel', 'energy'])

class MaxwellDemonSimulation:
    def __init__(self, num_particles=100, container_size=(10, 10)):
        self.num_particles = num_particles
        self.container_size = container_size
        self.particles = self._initialize_particles()
        self.partition_position = container_size[0] / 2
        
    def _initialize_particles(self):
        particles = []
        for i in range(self.num_particles):
            mass = np.random.uniform(1, 2)  # Different masses
            pos = np.array([np.random.uniform(0, self.container_size[0]),
                           np.random.uniform(0, self.container_size[1])])
            vel = np.random.uniform(-1, 1, 2)
            energy = 0.5 * mass * np.sum(vel**2)
            particles.append(Particle(i, mass, pos, vel, energy))
        return particles
    
    def calculate_thermodynamic_properties(self):
        left_particles = [p for p in self.particles if p.pos[0] < self.partition_position]
        right_particles = [p for p in self.particles if p.pos[0] >= self.partition_position]
        
        def calculate_properties(particles):
            if not particles:
                return 0, 0, 0
            masses = np.array([p.mass for p in particles])
            velocities = np.array([p.vel for p in particles])
            energies = np.array([p.energy for p in particles])
            
            temperature = np.mean(energies) / (1.5 * 1.38e-23)  # Simplified
            pressure = np.sum(masses * np.sum(velocities**2, axis=1)) / (2 * self.container_size[0])
            entropy = -np.sum(energies / np.sum(energies) * np.log(energies / np.sum(energies)))
            
            return temperature, pressure, entropy
        
        left_temp, left_press, left_ent = calculate_properties(left_particles)
        right_temp, right_press, right_ent = calculate_properties(right_particles)
        
        return {
            'left': {'temperature': left_temp, 'pressure': left_press, 'entropy': left_ent},
            'right': {'temperature': right_temp, 'pressure': right_press, 'entropy': right_ent}
        }
    
    def demon_logic(self, temperature_threshold):
        for particle in self.particles:
            if particle.pos[0] < self.partition_position:  # Left side
                if particle.energy > temperature_threshold:
                    particle.pos[0] = self.partition_position  # Move to right
            else:  # Right side
                if particle.energy < temperature_threshold:
                    particle.pos[0] = self.partition_position - 1e-3  # Move to left
    
    def run_simulation(self, steps=1000, temperature_threshold=1.0):
        temperatures = []
        pressures = []
        
        for step in range(steps):
            # Update particle positions
            for particle in self.particles:
                particle.pos += particle.vel
            
            # Apply demon logic
            self.demon_logic(temperature_threshold)
            
            # Calculate properties
            props = self.calculate_thermodynamic_properties()
            temperatures.append((props['left']['temperature'], props['right']['temperature']))
            pressures.append((props['left']['pressure'], props['right']['pressure']))
        
        return temperatures, pressures
```

```python
# Run simulation
sim = MaxwellDemonSimulation(num_particles=200)
temps, press = sim.run_simulation(steps=500, temperature_threshold=1.5)
```
